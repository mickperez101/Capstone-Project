## Formatting and Fixing File One - eSports Earnings 1998 -2023

#### The code listed below is the data being reformatted and cleaned up to make it easier to read for myself when it comes to numbers and data values in specific columns. The changes include : New Column Names, Converting Numbers into Cash Value, Shorten Genre Values, Dropping Old Rows, and Dropping specific values that are unnecessary in the data set.

```python
import pandas as pd
import matplotlib.pyplot as plt
import matplotlib
import numpy as np

eSports_earnings = pd.read_csv('archive/GeneralEsportData.csv')

## Display Max Rows in Columns

pd.set_option('display.max_rows', None)

## Column - TotalEarnings

## Converting Values to Cash Amount (to the nearest tenth)

eSports_earnings.loc[:,"Earnings"] = '$' + eSports_earnings['TotalEarnings'].map('{:,.2f}'.format)

## Column - Online Earnings

## Converting Values to Cash Amount (to the nearest tenth)

eSports_earnings.loc[:,"Digital_Earnings"] = '$' + eSports_earnings['OnlineEarnings'].map('{:,.2f}'.format)

## Column - Total Players

eSports_earnings.loc[:, "Tournaments"] = eSports_earnings['TotalTournaments']

## Column - Total Tournaments

eSports_earnings.loc[:, "Players"] = eSports_earnings['TotalPlayers']

## Column - Release Date 

eSports_earnings.rename(columns = {'ReleaseDate' : 'Year'}, inplace = True)

## Data Entry Changing Entries into Short Text

eSports_earnings['Genre'].replace({'Multiplayer Online Battle Arena':'Multiplayer','First-Person Shooter':'FPS', 
                     'Collectible Card Game':'TCG', 'Role-Playing Game' : 'RPG', 'Third-Person Shooter' : 'TPS',
                                  'Fighting Game' : 'Fighting', 'Puzzle Game' : 'Puzzle', 
                                   'Music / Rhythm Game' : 'Music'}, inplace = True)

## Dropping Old Columns or Rows // Code 

eSports_earnings.drop(columns = ['TotalEarnings'], axis =1, inplace = True)
eSports_earnings.drop(columns = ['OnlineEarnings'], axis =1, inplace = True)
eSports_earnings.drop(columns = ['TotalPlayers'], axis =1, inplace = True)
eSports_earnings.drop(columns = ['TotalTournaments'], axis =1, inplace = True)

## Drop any rows that have the value of 0 in every column, Unnecessary Values

eSports_earnings = eSports_earnings[(eSports_earnings['Earnings'] != 0) & (eSports_earnings['Digital_Earnings'] != 0) &
                                   (eSports_earnings['Tournaments'] != 0) & (eSports_earnings['Players'] != 0)]
```

```python
## Saving CSV For New CSV, With all changes above 

eSports_earnings.to_csv('archive/GeneralEsportData_update.csv', index=False)
```

## New Tournament CSV File

### Made a New File to Filter out certain values and reduce 600+ entries.

```python
## Reformatting for New Tournament CSV 

eSports_earnings = pd.read_csv('archive/GeneralEsportData.csv')

pd.set_option('display.max_rows', None)

eSports_earnings.loc[:,"Earnings"] = eSports_earnings['TotalEarnings'].map('{:,.2f}'.format)
eSports_earnings.loc[:,"Digital_Earnings"] = eSports_earnings['OnlineEarnings'].map('{:,.2f}'.format)
eSports_earnings.loc[:, "Tournaments"] = eSports_earnings['TotalTournaments']
eSports_earnings.loc[:, "Players"] = eSports_earnings['TotalPlayers']

eSports_earnings.rename(columns = {'ReleaseDate' : 'Year'}, inplace = True)

eSports_earnings['Genre'].replace({'Multiplayer Online Battle Arena':'Multiplayer','First-Person Shooter':'FPS', 
                     'Collectible Card Game':'TCG', 'Role-Playing Game' : 'RPG', 'Third-Person Shooter' : 'TPS',
                                  'Fighting Game' : 'Fighting', 'Puzzle Game' : 'Puzzle', 
                                   'Music / Rhythm Game' : 'Music'}, inplace = True)

eSports_earnings.drop(columns = ['TotalEarnings'], axis =1, inplace = True)
eSports_earnings.drop(columns = ['OnlineEarnings'], axis =1, inplace = True)
eSports_earnings.drop(columns = ['TotalPlayers'], axis =1, inplace = True)
eSports_earnings.drop(columns = ['TotalTournaments'], axis =1, inplace = True)

eSports_earnings = eSports_earnings[(eSports_earnings['Earnings'] != 0) & (eSports_earnings['Digital_Earnings'] != 0) &
                                   (eSports_earnings['Tournaments'] != 0) & (eSports_earnings['Players'] != 0)]

## Saving New CSV File

eSports_earnings.to_csv('archive/GeneralEsportData_tournament.csv', index=False)

```

### Making the New CSV File for Filters

```python
## Reading New CVS For Tournaments 

eSports_tournament = pd.read_csv('archive/GeneralEsportData_tournament.csv')

result = eSports_tournament.dtypes

## Makes the currency and removes any string values

def clean_currency(x):
    
    if isinstance(x,str):
        return(x.replace(',', ''))
    return(x)

eSports_tournament['Earnings'] = eSports_tournament['Earnings'].apply(clean_currency)
eSports_tournament['Digital_Earnings'] = eSports_tournament['Digital_Earnings'].apply(clean_currency)

## Rows Delete based on the Fitler, Less than 900 Tournaments

indexTournament = eSports_tournament[ (eSports_tournament['Tournaments'] <= 1500 ) ].index

eSports_tournament.drop(indexTournament, inplace = True)

## Convert to Numertic 

eSports_tournament['Earnings'] = pd.to_numeric(eSports_tournament['Earnings'])

## Decimal Places/Make it in the Millions  

eSports_tournament['Earnings'] = (eSports_tournament['Earnings'].astype(float)/1000000).round(2)
eSports_tournament['Digital_Earnings'] = (eSports_tournament['Digital_Earnings'].astype(float)/1000000).round(2)
```

```python
## Charts for Genres Per Game in the Data Set 

## Variables for the Chart (Genre per Game)

value = eSports_update['Genre'].value_counts()
genre_name = list(value.index)
genre_count = list(value.values)

## Chart Number One For Genre Per Game

matplotlib.style.use('ggplot')

genre_name.reverse()
genre_count.reverse()

plt.barh (genre_name, genre_count)
plt.title('Number Games for Each Genre')
plt.xlabel('Total Value')
plt.ylabel('Games Genre')


plt.show()
```

```python
import matplotlib.pyplot as plt
import matplotlib.pyplot as plt
import matplotlib.style as style

# Set the style to 'ggplot'
style.use('ggplot')

# Assuming eSports_update is your DataFrame, and 'Genre' is a column in it
alt_value = eSports_update['Genre'].value_counts()

# Extract genre names and corresponding counts
genre_nam = list(alt_value.index)
genre_num = list(alt_value.values)

# Calculate percentages separately
total = sum(genre_num)
percentages = [f'{(count / total * 100):.0f}%' for count in genre_num]

# Create a figure with a single subplot
fig, ax = plt.subplots(figsize=(10, 6))

# Plot the pie chart without percentage labels
wedges, texts, autotexts = ax.pie(genre_num, labels=None, autopct='', startangle=70)

# Equal aspect ratio ensures that the pie is drawn as a circle
ax.axis('equal')  

# Create a legend outside the pie chart with calculated percentage labels
legend_labels = [f'{label} ({percentage})' for label, percentage in zip(genre_nam, percentages)]
ax.legend(wedges, legend_labels, title="Genres", loc="center left", bbox_to_anchor=(1, 0, 0.5, 1))

# Display the plot
plt.show()

```
